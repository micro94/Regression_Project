### 가설 : 
1. 높은 점수를 가지는 조합이 있을것이다
2. 높은 점수를 가지는 챔피언이 존재할것이다
3. 높은 점수를 가지는 챔피언과 조합이 있을것이다

### 데이터 소개
    - 티어별 엑셀 데이터(5개)
        - TFT_Challenger_MatchData, TFT_Diamond_MatchData, TFT_GrandMaster_MatchData, TFT_Master_MatchData, TFT_Platinum_MatchData
    - columns 설명
        - 1) gameid: 게임 ID
        - 2) gameDuration: 게임 지속시간
        - 3) level: 플레이어 레벨(1~9)
        - 4) lastRound: 게임 죽기 전 라운드 수
        - 5) Ranked: 게임 순위(1~8위)
        - 6) ingameDuration: 플레이어별 게임 플레이 시간
        - 7) combination: 시너지 조합 
        - 8) champion: 플레이어에 사용된 챔피언 조합
    


### 종속 변수 :
